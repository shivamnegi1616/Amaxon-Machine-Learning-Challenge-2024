```python
# Function to extract and normalize entity value
def extract_answer(text, entity_name):

    # Clean up the text
    clean_text = re.sub(r'(?<=\d),(?=\d)', '', text)  # Remove commas between numbers
    clean_text = re.sub(r'\s+', ' ', clean_text)  # Normalize multiple spaces to one

    # Remove all brackets, commas, and @ symbols
    clean_text = re.sub(r'[()\[\]{}@,]', '', clean_text)

    # Optional: strip leading/trailing spaces
    clean_text = clean_text.split(' ')

    # Create a regex pattern to match values followed by units
    unit_variants = [re.escape(unit) for unit in unit_normalization_map.keys()]
    pattern = r'(\d+\.?\d*)\s*(' + '|'.join(unit_variants) + r')\b'

    # My Rules:
    for i in range(len(clean_text)):
        if clean_text[i] == entity_name:
            break

    clean_text.append(entity_name)


    clean_text = ' '.join(s for s in clean_text)

    # Search for all matches of the pattern in the clean text
    matches = re.findall(pattern, clean_text, re.IGNORECASE)

    # Check each match for validity against the entity_unit_map
    for value, unit in matches:
        normalized_unit = unit_normalization_map.get(unit.lower(), unit)
        if normalized_unit in entity_unit_map[entity_name]:
            return f"{float(value)} {normalized_unit}"

    return ""  # Return empty string if no valid match is found
```
